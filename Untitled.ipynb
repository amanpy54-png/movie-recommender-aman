```python
movies= movies[["id","title","genres","overview","keywords","cast","crew"]]
```

```python
# movies.head(3)
```

```python
# movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
# movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    Lst = []
    for i in ast.literal_eval(obj):
        Lst.append(i['name'])
    return Lst
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# movies.head(1)
```

```python
# movies["cast"][0]
def convert_cast(obj):
    Lst = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            Lst.append(i['name'])
            counter+=1
        else:
            break
    return Lst
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
# movies["crew"][0]
```

```python
import ast
def fetch_director(obj):
    Lst = []
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            Lst.append(i['name'])
            break
    return Lst
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# movies.head(2)
```

```python
# movies["overview"][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.lower().split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
# movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['crew'] + movies['keywords']
```

```python
new_movies = movies[["id", "title","tags"]].copy()
```

```python
# new_movies.head(1)
```

```python
# new_movies['tags'].head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
new_movies['tags'] = new_movies['tags'].apply(lambda x: x.lower())
```

```python
from nltk.stem import WordNetLemmatizer
import nltk

lemmatizer = WordNetLemmatizer()
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join([lemmatizer.lemmatize(word) for word in x.split()]))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
vectors = tfidf.fit_transform(new_movies['tags']).toarray()

feature_names = tfidf.get_feature_names_out()
print(feature_names[:20])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import pickle
similarity = cosine_similarity(vectors)
pickle.dump(similarity, open('similarity.pkl', 'wb'))
pickle.dump(new_movies.to_dict(), open('movie_dict.pkl', 'wb'))
```

```python
def recommend(movie_title):
    if movie_title not in new_movies['title'].values:
        print("Movie not found!")
        return

    movie_index = new_movies[new_movies['title'] == movie_title].index[0]
    distances = list(enumerate(similarity[movie_index]))
    distances = sorted(distances, reverse=True, key=lambda x: x[1])

    recommended = []
    for i, _ in distances:
        title = new_movies.iloc[i]['title']
        if title != movie_title and title not in recommended:
            recommended.append(title)
        if len(recommended) == 5:
            break

    print(f"Top 5 recommendations for '{movie_title}':")
    for title in recommended:
        print(title)
```

```python
recommend("Spider-Man 3"
)
```
